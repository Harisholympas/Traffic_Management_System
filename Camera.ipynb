```python
import cv2
from ultralytics import YOLO


model = YOLO(r"C:\projects\trafficmanagementsystem\yolov8n.pt")  


cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()


total_count = 0

while True:
    
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read frame.")
        break

    
    results = model(frame)  

    
    current_frame_count = 0

    
    for box in results[0].boxes.data:  
        x1, y1, x2, y2, conf, cls = box.tolist()  
        current_frame_count += 1  

        
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        label = f"Class: {int(cls)}, Confidence: {conf:.2f}"
        cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    
    total_count += current_frame_count

    
    cv2.putText(frame, f"Total Objects Detected: {total_count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

    
    cv2.imshow('YOLOv8 Real-Time Object Detection', frame)

    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


print(f"Total Objects Detected: {total_count}")


cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO
import time

model = YOLO(r"C:\projects\trafficmanagementsystem\yolov8n.pt")

cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

capture_interval = 110  # seconds
last_capture_time = time.time()

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read frame.")
        break

    current_time = time.time()
    if current_time - last_capture_time >= capture_interval:
        # Perform object detection on the captured frame
        results = model(frame)
        
        # Count objects in the current frame
        object_count = len(results[0].boxes.data)
        
        print(f"Objects detected in the current image: {object_count}")
        
        # Optionally, save the image
        cv2.imwrite(f"capture_{int(current_time)}.jpg", frame)
        
        # Reset the timer
        last_capture_time = current_time

    # Display the frame (optional, you can remove this if you don't want to see the video feed)
    cv2.imshow('YOLOv8 Periodic Object Detection', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
