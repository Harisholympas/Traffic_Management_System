```python
import cv2
from ultralytics import YOLO
import supervision as sv


model = YOLO(r"C:\projects\trafficmanagementsystem\yolov8n.pt")  


cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Initialize total count
total_count = 0

# Create an object to handle object counting
counter = sv.ObjectCounter()

while True:
    # Read a frame from the video capture
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read frame.")
        break

    # Perform object detection on the frame
    results = model(frame)

    # Extract detected boxes and classes
    detections = results[0].boxes.data

    # Update the counter with the current frame detections
    current_frame_count = counter.update(detections)

    # Draw bounding boxes and labels on the frame
    for box in detections:
        x1, y1, x2, y2, conf, cls = box.tolist()
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        label = f"Class: {int(cls)}, Confidence: {conf:.2f}"
        cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Update total count
    total_count += current_frame_count

    # Display total count on the frame
    cv2.putText(frame, f"Total Objects Detected: {total_count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

    # Show the frame with detections
    cv2.imshow('YOLOv8 Real-Time Object Detection', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


print(f"Total Objects Detected: {total_count}")


cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO
from supervision.video.dataclasses import VideoInfo
from supervision.video.sink import VideoSink
from tqdm import tqdm
import time
import pandas as pd

TARGET_VIDEO_PATH = "output_video.mp4"
EXCEL_PATH = "detection_counts.xlsx"

model = YOLO(r"C:\projects\trafficmanagementsystem\yolov8n.pt")

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS)) if int(cap.get(cv2.CAP_PROP_FPS)) > 0 else 30  
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

video_info = VideoInfo(
    width=width,
    height=height,
    fps=fps,
    total_frames=total_frames
)

total_count = 0
last_detection_time = time.time()  

detection_counts = []

with VideoSink(TARGET_VIDEO_PATH, video_info) as sink:
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error: Could not read frame.")
            break

        current_time = time.time()
        elapsed_time = current_time - last_detection_time

        if elapsed_time >= 10:
            results = model(frame)
            detections = results[0].boxes.data
            current_frame_count = len(detections)

            for box in detections:
                x1, y1, x2, y2, conf, cls = box.tolist()
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
                label = f"Class: {int(cls)}, Confidence: {conf:.2f}"
                cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

            total_count += current_frame_count
            detection_counts.append({'Frame': cap.get(cv2.CAP_PROP_POS_FRAMES), 'Objects Detected': current_frame_count})
            last_detection_time = current_time  

        cv2.putText(frame, f"Total Objects Detected: {total_count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
        sink.write_frame(frame)
        cv2.imshow('YOLOv8 Real-Time Object Detection', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

print(f"Total Objects Detected: {total_count}")

cap.release()
cv2.destroyAllWindows()


df = pd.DataFrame(detection_counts)
df.to_excel(EXCEL_PATH, index=False)
```
